```python
#horizontal plot

#import liabraries
import seaborn as sns
from numpy import median
import matplotlib.pyplot as plt

#load dataset

kashti = sns.load_dataset('titanic')
kashti
#draw a bar plot
sns.barplot(x='fare',y='class',hue='sex',data=kashti, estimator=median , saturation=1)
plt.show()
```

```python
#import liabraries
import seaborn as sns
import matplotlib.pyplot as plt

#load data
kashti = sns.load_dataset('titanic')
kashti
sns.barplot(x='fare',y='class',data=kashti,
            linewidth = 2, #bar line width
            facecolor=(1,0,0.5,1) #facecolor represents - RGBA (red, green, blue, alpha) where RGB are color intensities
                                #that can be defined as a value within range [0, 1] 
                                #and A controls the opacity of the facecolor i.e also a value within range [0, 1]
           ,errcolor='0' # bar color change from 0.1 to 1
           ,edgecolor = '0.3' #bar edge color from 0.1 to 1
           )
plt.show()
```
